```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Assuming 'data' is your DataFrame and has been preprocessed (features scaled)
features = df_selected.drop('diagnosis', axis=1)  # Excluding target variable
features_scaled = StandardScaler().fit_transform(features)  # Standardizing the features

# Applying PCA
pca = PCA(n_components=5)  # Adjust components based on explained variance
principal_components = pca.fit_transform(features_scaled)
explained_variance = pca.explained_variance_ratio_

print(f"Explained Variance by Principal Components: {explained_variance}")

import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Assuming 'data' is preprocessed and standardized
pca = PCA()
pca.fit(features_scaled)
explained_variance = pca.explained_variance_ratio_

plt.figure(figsize=(8, 6))
plt.plot(range(1, len(explained_variance)+1), explained_variance, marker='o')
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Variance Explained')
plt.xticks(range(1, len(explained_variance)+1))
plt.grid(True)
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Assuming 'data' is preprocessed and standardized
pca = PCA()
pca.fit(features_scaled)
explained_variance = pca.explained_variance_ratio_

cumulative_variance = np.cumsum(explained_variance)

plt.figure(figsize=(8, 6))
plt.plot(range(1, len(cumulative_variance)+1), cumulative_variance, marker='o')
plt.title('Cumulative Explained Variance')
plt.xlabel('Principal Component')
plt.ylabel('Cumulative Variance Explained')
plt.xticks(range(1, len(cumulative_variance)+1))
plt.grid(True)
plt.show()
```
